### What is numpy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.


At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

# Numpy array vs python list (speed and memory)

```python
# Numpy array vs list (speed)

# python list

a = [x for x in range(1,10000000)]
b = [x for x in range(1,10000000)]
result = []
start = time.time()
for i in range(len(a)):
  result.append(a[i] + b[i])
print(f"Time taken for the python list : {time.time() - start} sec")
```

```python
#numpy array

np_a = np.array(a)
np_b = np.array(b)
start = time.time()
np_result = np_a + np_b
np_result
print(f"Time taken for the Numpy array : {time.time() - start} sec")
```

```python
#numpy array vs list(memory)

print(f"size of python list {sys.getsizeof(a)}")
print(f"size of numpy array {sys.getsizeof(np_a)}")
```

**random module**



*   **random.rand()** : Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
*   **random.randint()**: Return random integers from low (inclusive) to high (exclusive).

    Return random integers from the “discrete uniform” distribution of the specified dtype in the “half-open” interval [low, high). If high is None (the default), then results are from [0, low).

**random.uniform():**Draw samples from a uniform distribution.
    
  Samples are uniformly distributed over the
    half-open interval [low, high) (includes low, but excludes high). In other words, any value within the given interval is equally likely to be drawn by uniform.

**random.normal(loc=0.0, scale=1.0, size=None):**Draw
    random samples from a normal (Gaussian) distribution.

**random.randn():**return a sample (or samples) from
    the “standard normal” distribution.

**random.random_sample(size=None):**Return random
    floats in the half-open interval [0.0, 1.0).

**random.shuffle():**Modify a sequence in-place by
    shuffling its contents.
    This function only shuffles the array along the first axis of a multi-dimensional array. The order of sub-arrays is changed but their contents remains the same.

```python
arr = np.arange(1,13).reshape(4,3)
print(arr)
print('\n')
print("-----------mean---------------")
print(np.mean(arr,axis=0))#axis=0-->columns & axis=1-->row
print('\n')
print('------------median-------------')
print(np.median(arr,axis=1))
print('\n')
print('------------variance-------------')
print(np.var(arr,axis=1))
print('\n')
print('------------standard deviation-------------')
print(np.std(arr,axis=1))
print('\n')
print('------------min/max-------------')
print(np.max(arr,axis=1))
print('\n')
print('------------range-------------')
print(np.ptp(arr,axis=1))
print('\n')
print('------------percentiles-------------')
print(np.percentile(arr,axis=0,q=80))
print('\n')
print('------------quantiles-------------')
print(np.quantile(arr,axis=0,q=0.8))
print('\n')
print('------------histogram-------------')
print(np.histogram(arr))
print('\n')
```

**noraml indexing and slicing**


```python

#normal indexing : a[row_index][col_index]
#first row first col element
a[0][0]
print(a[2][3])
print(a[2,3])

#normal slicing : a[start:stop:step] --- > works on rows
a[1:4:2]
a[:3:1]

# advanced slicing
# array[row_slice , col_slice] -- > array[row_start:row_end-1:row_step , col_start:col_end-1:col_step]
print(a)
a[2:,2:]

```

**Fancy indexing**

```python
#used when only required some col or rows to select
a[:,[1,2]]
a[[1,2],:]
```

**Boolean Indexing**

```python
print(a)
a[a>50]
a[a%2 == 0] #even numbers
a[a%8==0] #mul of 8

#we can use &,| operators to chain the operations
# number which is greater than 40 or divisible by 3

print(a[(a>40) | (a%3==0)])
```

# stcking and slitting

```python
arr1 = np.random.randint(40,size=(4,4))
arr2 = np.random.randint(10,size=(4,4))

print(arr1)
print(arr2)
```

**horizontal stacking**

```python
#stacks the arrays side by side

h_res=np.hstack((arr1,arr2))
print(h_res)
```

**vertical stacking**

```python
#stacks the arrays one upon the other
v_res=np.vstack((arr1,arr2))
print(v_res)
```

**horizontal splitting**

```python
#rows remains same splits the columns
print(h_res)
print('\n')
print(v_res)
print('\n')
h_split = np.hsplit(h_res,2)
print(h_split)
print('\n')
v_split = np.hsplit(v_res,2)
print(v_split)

```

**vertical splitting**

```python
#columns remains same but rows get effected
print(h_res)
print('\n')
print(v_res)
print('\n')
h_split = np.vsplit(h_res,2)
print(h_split)
print('\n')
v_split = np.vsplit(v_res,2)
print(v_split)
```

# Broadcasting

**Broadcasting Rules**

**1. Make the two arrays have the same number of dimensions.**<br>
- If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.<br>

**2. Make each dimension of the two arrays the same size.**<br>
- If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other array.
- If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.

**np.sort()**: sorts the elements of the array

```python
#descending
np.random.seed(100)
arr1 = np.random.randint(34,size=(6,5))
print(f"Array : \n{arr}")
print('\n')
sorted_arr = np.sort(arr1,axis=1)[::-1]
print(f"sorted_array along row_wise : \n{sorted_arr}")
print('\n')
sorted_arr1 = np.sort(arr1,axis=0)[::-1]
print(f"sorted_array along col_wise : \n{sorted_arr1}")
```
